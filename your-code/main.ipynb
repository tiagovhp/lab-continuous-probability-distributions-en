## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# your code here
import random
def generate_unif_random_nrs(bottom, ceiling, count):
    random_nrs = []
    for i in range(count):
        random_nrs.append(random.uniform(bottom, ceiling))
    return random_nrs

```

1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50

```python
import random

def generate_normal_random_nrs(mu, sigma, count):
    return [random.gauss(mu, sigma) for _ in range(count)]

```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
# 1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10.
def generate_exponential_samples(mean,size):
    
    sample = np.random.exponential(mean, size)
    return sample

#2. Use the function to generate two number sequences with the size of 10 and 100.
seq_1 = pd.DataFrame(generate_exponential_samples(10,10),columns=['value'])

seq_2 = pd.DataFrame(generate_exponential_samples(10,100),columns=['value'])




```

```python

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Plot the histogram for the first sample
plt.subplot(1, 2, 1)
sns.histplot(seq_1['value'], bins=100, kde=True, color='blue')
plt.title('Histogram of 10 Samples (Exponential Mean = 10)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the histogram for the second sample
plt.subplot(1, 2, 2)
sns.histplot(seq_2['value'], bins=100, kde=True, color='orange')
plt.title('Histogram of 100 Samples (Exponential Mean = 10)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()
plt.show()
```

How are the two distributions different?

The mean changes, so the distribution changes as well.
